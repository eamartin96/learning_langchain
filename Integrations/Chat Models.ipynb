### Chat Model Initialization

```python
from langchain_openai import ChatOpenAI

chat_model = ChatOpenAI()
```

### Chat Model Invocation

```python
chat_model.invoke("Hello how are you?")
```

By default the model used is `gpt-3.5-turbo`. We can specify the model in the `ChatOpenAI` instantiation

```python
chat_model = ChatOpenAI(
    model = "gpt-4"
)

chat_model.invoke("Hello how are you?")
```
